```python
import tkinter as tk
from tkinter import messagebox
from bardapi import Bard
import os

# Function to get the answer from Bard API
def get_answer():
    try:
        api_key = api_key_entry.get()
        os.environ['_BARD_API_KEY'] = api_key
        
        field = field_entry.get()
        question = question_entry.get()
        
        bard = Bard()
        response = bard.get_answer(question)
        
        if 'content' in response:
            answer_text.delete(1.0, tk.END)  # Clear the text box before inserting new content
            answer_text.insert(tk.END, response['content'])
        else:
            messagebox.showerror("Error", "No content received. Please check your API key and question.")
    except Exception as e:
        messagebox.showerror("Error", f"An error occurred: {e}")

# Function to toggle API key visibility
def toggle_key_visibility():
    if api_key_entry.cget('show') == '':
        api_key_entry.config(show='*')
        toggle_button.config(text="Show")
    else:
        api_key_entry.config(show='')
        toggle_button.config(text="Hide")

# Setting up the main window
app = tk.Tk()
app.title("InsightHub")
app.geometry("600x600")
app.configure(bg="#e0f7fa")  # Background color

# Frame for padding and layout
frame = tk.Frame(app, bg="#e0f7fa", padx=20, pady=20)
frame.pack(expand=True)

# API Key Entry
tk.Label(frame, text="Enter your API Key:", bg="#e0f7fa", font=("Helvetica", 12)).pack(anchor='w')
api_key_entry = tk.Entry(frame, show='*', width=40, font=("Helvetica", 12))
api_key_entry.pack()

# Toggle visibility button for API Key
toggle_button = tk.Button(frame, text="Show", command=toggle_key_visibility, bg="#80deea", fg="black", relief="flat")
toggle_button.pack(pady=5)

# Field Entry
tk.Label(frame, text="Enter your Field:", bg="#e0f7fa", font=("Helvetica", 12)).pack(anchor='w', pady=(10, 0))
field_entry = tk.Entry(frame, width=50, font=("Helvetica", 12))
field_entry.pack()

# Question Entry
tk.Label(frame, text="Enter your Question:", bg="#e0f7fa", font=("Helvetica", 12)).pack(anchor='w', pady=(10, 0))
question_entry = tk.Entry(frame, width=50, font=("Helvetica", 12))
question_entry.pack()

# Get Answer Button
get_answer_button = tk.Button(frame, text="Get Answer", command=get_answer, 
                              bg="#4dd0e1", fg="white", font=("Helvetica", 12, "bold"), 
                              relief="flat", padx=10, pady=5)
get_answer_button.pack(pady=15)

# Answer Display Box
answer_text = tk.Text(frame, wrap='word', width=60, height=10, font=("Helvetica", 12), 
                      bg="#ffffff", fg="black", relief="solid", bd=1)
answer_text.pack()

# Run the app
app.mainloop()
```
